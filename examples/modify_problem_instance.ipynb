# Modify an existing `problem_instance` JSON file

This script is used to update/change some fields in a `problem_instance` file and write it back out.  This will only be used during preliminary compilation of `problem_instance` files for corrections or schema updates and when find/replace isn't sufficient.

```python
import json
import requests
import pprint

import sys
sys.path.append("../")


# additional package(s):
import jsonschema
```

```python
# example file:
file_path = "../problem_instances/problem_instance.e36d77e9-e467-49dc-8c87-973fd6776f18.json"


# load data from file as a Python dictionary object:
with open(file_path, "r") as json_file:
    problem_instance = json.load(json_file)

pprint.pprint(problem_instance)
```

```python



# Access and modify some property (example)
print(problem_instance["instance_data"][2]["requirements"]["time_limit_seconds"])

# modify the property if you need to.
problem_instance["instance_data"][2]["requirements"]["time_limit_seconds"] = 86401
 
print(problem_instance["instance_data"][2]["requirements"]["time_limit_seconds"])
```

```python
# write the modified problem_instance back out to the file


# STRONGLY RECOMMEND backing up problem_instance files and ALSO writing modified problem_instance files out to a temporary directory for inspection first.

output_json_file_path = "../archive/temp_modified_instances/problem_instance." + problem_instance["problem_instance_uuid"] + ".json"
with open(output_json_file_path, "w") as json_file:
    json.dump(problem_instance, json_file)
```
